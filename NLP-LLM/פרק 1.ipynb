```python
import transformers
from transformers import pipeline
```

```python
print(f"Version: {transformers.__version__}")
```

```python
classifier = pipeline("sentiment-analysis")

classifier("I've been waiting for a HuggingFace course my whole life.")
```

```python
classifier([
"I've been waiting for a HuggingFace course my whole life.",
"I hate this so much!"
])
```

```python

classifier = pipeline("zero-shot-classification")

classifier(
"This is a course about the Transformers library",
candidate_labels=["education", "politics", "business"]
)
```

```python
generator = pipeline("text-generation", max_length=1000)

generator("In this course we will teach you how to")
```

```python
generator = pipeline("text-generation", model="HuggingFaceTB/SmolLM2-360M", num_return_sequences=1, max_length=50)

generator("In this course we will teach you how to")
```

```python
generator = pipeline("text-generation")

generator("יום אחד הלכתי")
```

```python
unmasker = pipeline("fill-mask")

unmasker(
"This course will teach you all about <mask> models."
)
```

```python
ner = pipeline("ner", model="dslim/bert-base-NER", aggregation_strategy="simple")

text = "My name is Sylvain and I work at Hugging Face in Brooklyn"

results = ner(text)
```

```python
results
```

```python
question_answerer = pipeline("question-answering")

context = "My name is Sylvain and I work at Hugging Face in Brooklyn"
question = "Where do I work?"
question_answerer(question=question, context=context)
```

```python
translator = pipeline(
    "translation_en_to_he",
    model="Helsinki-NLP/opus-mt-en-he"
)

text = "Hello, how are you?"
translator(text)
```

```python
summarizer = pipeline("text2text-generation", model="google/flan-t5-small")

text = """
Hugging Face is a company based in New York that develops tools for building applications using machine learning.
It is widely known for its Transformers library, which provides thousands of pretrained models for natural language processing tasks.
"""
summarizer(f"summarize: {text}", max_length=20, min_length=2)
```
