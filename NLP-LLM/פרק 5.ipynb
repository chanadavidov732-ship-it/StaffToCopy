```python
# פתרון פשוט ל-mismatch:
# מוחקים את העמודות המקוריות ומשאירים רק את הפלט החדש של tokenizer
tokenized_dataset = drug_dataset.map(
    tokenize_and_split,
    batched=True,
    remove_columns=drug_dataset["train"].column_names
)
```

```python
# משווים בין מספר הדוגמאות אחרי הפיצול לבין הדאטה המקורי
len(tokenized_dataset["train"]), len(drug_dataset["train"])
```

```python
# פתרון מתקדם:
# שומרים גם את העמודות המקוריות על ידי שכפול שלהן לפי ה-chunks החדשים
def tokenize_and_split(examples):
    result = tokenizer(
        examples["review"],
        truncation=True,
        max_length=128,
        return_overflowing_tokens=True
    )

    # מיפוי: כל chunk חדש יודע מאיזו דוגמה מקורית הוא הגיע
    sample_map = result.pop("overflow_to_sample_mapping")

    # משכפלים את כל הערכים המקוריים לפי המיפוי
    for key, values in examples.items():
        result[key] = [values[i] for i in sample_map]

    return result
```

```python
import time
import math
from pathlib import Path
import pandas as pd
from tqdm.notebook import tqdm

# פונקציה שמורידה issues מ־GitHub ושומרת כ־JSONL
def fetch_issues(
    owner="huggingface",
    repo="datasets",
    num_issues=10_000,
    rate_limit=5_000,
    issues_path=Path("."),
):
    if not issues_path.is_dir():
        issues_path.mkdir(exist_ok=True)

    batch = []
    all_issues = []
    per_page = 100  # מספר issues בכל בקשה
    num_pages = math.ceil(num_issues / per_page)
    base_url = "https://api.github.com/repos"

    for page in tqdm(range(num_pages)):
        # מביאים גם issues וגם pull requests
        query = f"issues?page={page}&per_page={per_page}&state=all"
        issues = requests.get(f"{base_url}/{owner}/{repo}/{query}", headers=headers)
        batch.extend(issues.json())

        # אם עברנו את ה־rate limit — מחכים שעה
        if len(batch) > rate_limit and len(all_issues) < num_issues:
            all_issues.extend(batch)
            batch = []
            print(f"Reached GitHub rate limit. Sleeping for one hour ...")
            time.sleep(60 * 60 + 1)

    all_issues.extend(batch)

    # ממירים ל־DataFrame
    df = pd.DataFrame.from_records(all_issues)

    # שומרים לקובץ JSONL
    df.to_json(f"{issues_path}/{repo}-issues.jsonl", orient="records", lines=True)

    print(
        f"Downloaded all the issues for {repo}! Dataset stored at {issues_path}/{repo}-issues.jsonl"
    )
```

```python
# מסננים תגובות קצרות מדי
comments_dataset = comments_dataset.filter(lambda x: x["comment_length"] > 15)
comments_dataset
```

```python
# פונקציה שמחברת title + body + comment לטקסט אחד
def concatenate_text(examples):
    return {
        "text": examples["title"]
        + " \n "
        + examples["body"]
        + " \n "
        + examples["comments"]
    }

comments_dataset = comments_dataset.map(concatenate_text)
```

```python
# טוענים מודל ו-tokenizer ליצירת embeddings
from transformers import AutoTokenizer, AutoModel

model_ckpt = "sentence-transformers/multi-qa-mpnet-base-dot-v1"
tokenizer = AutoTokenizer.from_pretrained(model_ckpt)
model = AutoModel.from_pretrained(model_ckpt)
```

```python
# מעבירים את המודל ל-GPU
import torch

device = torch.device("cuda")
model.to(device)
```

```python
# פונקציה שלוקחת את ה-[CLS] token כייצוג המשפט
def cls_pooling(model_output):
    return model_output.last_hidden_state[:, 0]
```

```python
# פונקציה שמייצרת embeddings לטקסטים
def get_embeddings(text_list):
    encoded_input = tokenizer(
        text_list, padding=True, truncation=True, return_tensors="pt"
    )
    encoded_input = {k: v.to(device) for k, v in encoded_input.items()}
    model_output = model(**encoded_input)
    return cls_pooling(model_output)
```

```python
# בדיקה: יוצרים embedding לטקסט אחד
embedding = get_embeddings(comments_dataset["text"][0])
embedding.shape
```

```python
# מוסיפים עמודת embeddings לכל הדאטה
embeddings_dataset = comments_dataset.map(
    lambda x: {"embeddings": get_embeddings(x["text"]).detach().cpu().numpy()[0]}
)
```
